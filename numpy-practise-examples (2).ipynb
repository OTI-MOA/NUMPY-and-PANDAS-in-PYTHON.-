```python
asset1 = np.random.normal(0.001, 0.02, 1000)
asset2 = np.random.normal(0.001, 0.03, 1000)

w1 = 0.6
w2 = 0.4

cov_matrix = np.cov(asset1, asset2)

portfolio_variance = (
    w1**2 * cov_matrix[0,0] +
    w2**2 * cov_matrix[1,1] +
    2*w1*w2 * cov_matrix[0,1]
)

portfolio_volatility = np.sqrt(portfolio_variance)

print(portfolio_volatility)
```

```python
import numpy as np

np.random.seed(42)
initial = 10000
days = 252
sims = 1000

# STOCK A: "The Steady Climber" (Strong Motor, Small Waves)
# Mean = 0.0015 (Higher), Volatility = 0.01 (Lower)
returns_A = np.random.normal(0.0015, 0.01, (days, sims))
final_A = initial * np.cumprod(1 + returns_A, axis=0)[-1]

# STOCK B: "The Wild Rider" (Weak Motor, Huge Waves)
# Mean = 0.0005 (Lower), Volatility = 0.04 (Higher)
returns_B = np.random.normal(0.0005, 0.04, (days, sims))
final_B = initial * np.cumprod(1 + returns_B, axis=0)[-1]

# Calculate Probability of Loss for both
prob_loss_A = np.mean(final_A < initial)
prob_loss_B = np.mean(final_B < initial)

print(f"Stock A (Steady) - Prob of Loss: {prob_loss_A:.2%}")
print(f"Stock B (Wild)   - Prob of Loss: {prob_loss_B:.2%}")
```
