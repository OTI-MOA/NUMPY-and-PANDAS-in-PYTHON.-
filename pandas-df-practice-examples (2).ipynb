```python
import numpy as np

dates = pd.date_range(start="2024-01-01", periods=5)

df = pd.DataFrame({
    "date": dates,
    "returns": np.random.normal(0.001, 0.02, 5)
})

df.set_index("date", inplace=True)

print(df)
```

inplace=True tells the code not to create a new column index

```python
df["rolling_mean"] = df["returns"].rolling(window=3).mean()
df
```

calculates the value of 3day rolling window.
0 and 1 is NaN because it needs a 3day return value to cal the mean

```python
slice_to = df["2024-01-02": "2024-01-04"]
slice_to
```

```python
df["yesterdate"] = df["returns"].shift(1)
df
```

It shifts yesterday value to today. this is good for predicting model.this tells the computer to look at what happened yesterday to predict what might happen today.

```python
daily_returns = df["returns"] - df["yesterdate"]
print(daily_returns)
```

```python
weekly_returns = df["returns"].resample("W").mean()
print(weekly_returns)
```

**Time Series**

```python
df["rolling_mean"] = df["returns"].rolling(window=3, min_periods=1).mean()
df
```

day1 give you day1 value while day2 give average of day1 and 2 and so on.

By adding min_periods=1, you are telling Pandas: "Don't wait for a full 3-day window; give me whatever average you can calculate starting from Day 1."

```python
df["rolling_vol"] = df["returns"].rolling(3).std()
df
```

This code calculates a 3-Day Rolling Volatility. While the rolling mean tells you the trend, the rolling Standard Deviation (std) tells you how much the "heartbeat" of the market is speeding up or slowing down.


```python
df["expanding_val"] = df["returns"].expanding().mean()
df
```

this cal the average val of everyday except day1

```python
np.random.seed(42)

df = pd.DataFrame({
    "date": pd.date_range("2024-01-01", periods=10000, freq="h"),
    "amount": np.random.randint(1000,100000,10000),
    "currency": np.random.choice(["NGN","USD","EUR"],10000),
    "risk_score": np.random.randint(1,11,10000)
})

df.set_index("date", inplace=True)

monthly_volume = df.resample("ME")["amount"].sum()

monthly_high_risk = df.resample("ME")["risk_score"].apply(
    lambda x: (x > 7).mean()
)

top_risk_months = monthly_high_risk.sort_values(ascending=False).head(3)

print(monthly_volume)
print(monthly_high_risk)
print(top_risk_months)
```

This code creates a massive Transaction Monitoring System that analyzes over a year of hourly data to find the most dangerous months for a business.
Line-by-Line Explanation
1. pd.date_range(..., freq="H")
What it does: Generates 10,000 consecutive hours (about 13.8 months) of timestamps.
Real-Life Example: This mimics a Global Digital Bank. Unlike a local shop that opens 9-to-5, a digital bank processes payments 24/7, every single hour.
2. monthly_volume = df.resample("M")["amount"].sum()
What it does: Groups the 10,000 hours into months and adds up all the transaction amounts.
Real-Life Example: The Revenue Report. This tells the CEO how much total cash moved through the platform in January vs. February.
3. monthly_high_risk = df.resample("M")["risk_score"].apply(lambda x: (x > 7).mean())
What it does: For each month, it calculates what percentage of transactions had a risk score of 8, 9, or 10.
Real-Life Example: The Fraud Trend Tracker. It doesn't look at the money; it looks at the behavior. It tells you, "In March, 30% of our traffic looked suspicious."
4. top_risk_months = monthly_high_risk.sort_values(ascending=False).head(3)
What it does: Sorts the months from highest risk to lowest and picks the top 3 "worst" months.
Real-Life Example: The Incident Audit. If a bank gets attacked by hackers, the compliance team uses this to identify exactly which months had the highest density of red flags so they can investigate those specific time periods.


```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 5000

df = pd.DataFrame({
    "transaction_id": range(1, n+1),
    "customer_id": np.random.randint(1000, 1100, n),
    "date": pd.date_range("2024-01-01", periods=n, freq="h"),
    "amount": np.random.randint(1000, 200000, n),
    "currency": np.random.choice(["NGN","USD","EUR"], n),
    "risk_score": np.random.randint(1, 11, n)
})

df["is_fraud"] = (
    (df["amount"] > 150000) &
    (df["risk_score"] > 8)
).astype(int)
df
```

This code builds a Fraud Detection Filter. It’s the logic used by banks to decide which transactions should be instantly blocked or flagged for manual review.
Line-by-Line Explanation
1. n = 5000
What it does: Sets the scale to 5,000 transactions.
Real-Life Example: This is a "mid-sized" batch for a fintech—roughly what a busy payment gateway processes in an hour.
2. df = pd.DataFrame({...})
What it does: Creates a detailed ledger including Who (customer_id), When (date), How much (amount), and How safe (risk_score).
Real-Life Example: This is the Core Banking Ledger. It captures every detail needed for an audit trail.
3. df["is_fraud"] = ((df["amount"] > 150000) & (df["risk_score"] > 8)).astype(int)
What it does: This is the "Decision Engine." It looks for two specific red flags happening at once:
The amount is very high (> ₦150k).
The risk score is critical (9 or 10).
.astype(int): Converts True/False into 1 (Fraud) and 0 (Safe).
Real-Life Example: This is a Hard Rule. Imagine a security guard who only stops people if they are carrying a heavy bag and acting nervous. If only one is true, they pass; if both are true, they are stopped.

> This code is your "Pre-Flight Check." In a fintech audit, you run these commands to ensure your fraud detection rules actually worked and to see how many "criminals" you caught.
Line-by-Line Explanation
1. df.head()
What it does: Shows the first 5 rows of your data.
Real-Life Example: You're taking a quick look at the top of the pile of receipts to make sure the columns like is_fraud and amount actually exist and look correct.
2. df.info()
What it does: Checks for missing data and identifies data types.
Real-Life Example: You're checking the Health of the File. If the date column is listed as "object" (text) instead of "datetime," your time-based fraud filters might fail.
3. df.describe()
What it does: Gives you the "Stats." You see the average transaction amount, the minimum, and the maximum.
Real-Life Example: You look at the Maximum Amount. If your fraud rule is for ₦150k but the max transaction is ₦200M, you realize your rule might be too small and you're missing the "big fish."
4. df["is_fraud"].value_counts()
What it does: Counts how many 0s (Safe) and 1s (Fraud) are in your list.
Real-Life Example: The Daily Scorecard. It tells you exactly how many transactions were flagged.
If you see 4,950 zeros and 50 ones, your system is blocking 1% of traffic.
If you see 0 ones, your fraud rule is too strict and isn't catching anything!

```python
df["amount_band"] = pd.cut(
    df["amount"],
    bins=[0,50000,100000,150000,200000],
    labels=["Low","Medium","High","Very High"]
)
df.groupby("amount_band")["is_fraud"].mean()
```

This code performs Risk Segmentation. It groups your transactions into buckets based on their size and then calculates the "danger level" for each bucket.
Line-by-Line Explanation
1. pd.cut(df["amount"], bins=[...], labels=[...])
What it does: It takes the continuous numbers in the "amount" column and categorises them into four discrete folders:
₦0 – ₦50k 
 Low
₦50k – ₦100k 
 Medium
₦100k – ₦150k 
 High
₦150k – ₦200k 
 Very High
Real-Life Example: Think of a Bank's Withdrawal Limit. Instead of looking at every specific Naira, the bank treats you differently if you withdraw ₦10k (no ID needed) versus ₦180k (requires extra verification).
2. df.groupby("amount_band")["is_fraud"].mean()
What it does: It looks at each folder (Low, Medium, High, Very High) and calculates the percentage of transactions inside that folder that were marked as is_fraud.
Real-Life Example: You are asking the computer: "In which of these four folders are the criminals hiding?"


```python
customer_risk = df.groupby("customer_id").agg({
    "amount": "mean",
    "risk_score": "mean",
    "is_fraud": "mean"
})

customer_risk.sort_values("is_fraud", ascending=False).head()
```

This code creates a "Suspect Watchlist." It shifts the focus from individual transactions to the actual people behind them, identifying which customers are consistently triggering your fraud alerts.
Line-by-Line Explanation
1. df.groupby("customer_id").agg({...})
What it does: It gathers every transaction made by a specific person and calculates their "Profile."
amount: mean: Their average spending habit.
risk_score: mean: Their average "danger level" over time.
is_fraud: mean: The percentage of their total transactions that were flagged as fraud.
Real-Life Example: This is Behavioral Profiling. Instead of judging a customer by one mistake, you are looking at their entire history with your bank.
2. .sort_values("is_fraud", ascending=False).head()
What it does: It puts the most suspicious people at the very top of the list.
Real-Life Example: This is the Top 5 Most Wanted list for your compliance team.
